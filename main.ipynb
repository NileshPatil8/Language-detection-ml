First we have to convert the dataset from a string type into an array Then we need to convert that array into numbers.

```python
X = np.array(df["Text"])
y = np.array(df["language"])
```

```python
X
```

```python
y
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

CountVectorizer is based on the Bag of Words model. It converts lines of text into individual words for example, it picks one word, then another, and so on and assigns a unique number to each distinct word

```python
cv = CountVectorizer()
X = cv.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test =train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
y_test
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

```python
user = input("Enter text")
data= cv.transform([user]).toarray()
output = model.predict(data)
print(output)
```
